This program finds out a particular string('#love' in this case) in recent public tweets (count upto 100 in this case).

```python
import twitter
import json

#Get your keys and tokens from the Twitter Dev Console 
CONSUMER_KEY = "XXXXXXXXXXXXXXXXXXXXXXXXXXXX"
CONSUMER_SECRET = "XXXXXXXXXXXXXXXXXXXXXXXXXXXX"
OAUTH_TOKEN = "XXXXXXXXXXXXXXXXXXXXXXXXXXXX"
OAUTH_TOKEN_SECRET = "XXXXXXXXXXXXXXXXXXXXXXXXXXXX"

auth = twitter.oauth.OAuth(OAUTH_TOKEN,OAUTH_TOKEN_SECRET,CONSUMER_KEY,CONSUMER_SECRET)

twitter_api = twitter.Twitter(auth=auth)

print(twitter_api)
```

```python
file = open(r'Files/twitserach.txt','w')

str = '#love'

ctr = 100

search_results =  twitter_api.search.tweets(q=str, count = ctr)
print(json.dumps(search_results, indent=1))
file.write('\n search_results : \n' +json.dumps(search_results, indent = 1))
input('Press ENTER to Continue')

statuses = search_results['statuses']
print(json.dumps(statuses, indent = 1))
print(len(statuses))
file.write('\n statuses: \n' +json.dumps(statuses, indent = 1))
input('Press ENTER To CONTINUE')

```

```python
print('# Show one sample search result by slicing the list...')

print(json.dumps(statuses[0], indent=1))

file.write('\n statuses[0] : \n' +json.dumps(statuses[0], indent = 1))
for i in range(len(statuses)):
    try:
        print(i,statuses[i]['text'])
    except UnicodeEncodeError:
        pass

file.close
```
