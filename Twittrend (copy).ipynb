This program helps us to evaluate latest twitter trends in World, United States & India

```python
import twitter
import json

# Get your keys and tokens from the Twitter Dev Console
CONSUMER_KEY = "XXXXXXXXXXXXXXXXXXXXXXXX"
CONSUMER_SECRET = "XXXXXXXXXXXXXXXXXXXXXXXX
OAUTH_TOKEN = "XXXXXXXXXXXXXXXXXXXXXXXX"
OAUTH_TOKEN_SECRET = "XXXXXXXXXXXXXXXXXXXXXXXX"

auth = twitter.oauth.OAuth(OAUTH_TOKEN,OAUTH_TOKEN_SECRET,CONSUMER_KEY,CONSUMER_SECRET)

twitter_api = twitter.Twitter(auth=auth)

print(twitter_api)
```

```python
WORLD_WOE_ID= 1
US_WOE_ID= 23424977
INDIA_WOE_ID= 23424848

world_trends = twitter_api.trends.place(_id=WORLD_WOE_ID)
us_trends = twitter_api.trends.place(_id=US_WOE_ID)
india_trends = twitter_api.trends.place(_id=INDIA_WOE_ID)

file = open(r'Files/twittrend.txt','w')

print('======================WORLD TRENDS==================')
print(json.dumps(world_trends,indent = 1))
input('Press Enter Key to Continue')
file.write("world_trends : \n" +json.dumps(world_trends,indent=1)+'\n')

```
